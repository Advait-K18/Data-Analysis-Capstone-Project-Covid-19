```python
plt.figure(figsize=(12,6))
sns.relplot(data=covid,x="Age Group",y="Total Deaths",kind='line')
plt.xticks(rotation=45)
plt.show()
```

```python
# Aggregate deaths over time
time_series = covid.groupby('End Date')['COVID-19 Deaths'].sum()
time_series
```

```python
#COVID-19 deaths over time
plt.figure(figsize=(12,6))
plt.plot(time_series, marker='o', linestyle='-', color='red')
plt.title("COVID-19 Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.grid()
plt.show()
```

```python
# Visualizing COVID-19 deaths by State
plt.figure(figsize=(15, 6))
top_states = covid.groupby("State")["COVID-19 Deaths"].sum().sort_values(ascending=False).head(15)
sns.barplot(x=top_states.index, y=top_states.values, palette="viridis")
plt.title("Top 15 States with Highest COVID-19 Deaths")
plt.xlabel("State")
plt.ylabel("Total COVID-19 Deaths")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

The bar chart displays the top 15 states with the highest total COVID-19 deaths. Some states have significantly higher numbers, likely due to population density and healthcare factors.

```python
top_5_states = covid.groupby("State")["COVID-19 Deaths"].sum().sort_values(ascending=False).head(5)
```

```python
# Replotting COVID-19 Death Trends Over Time for Top 5 States Separately

fig, axes = plt.subplots(3, 2, figsize=(15, 12))  
axes = axes.flatten()  

for i, state in enumerate(top_5_states):
    ax = axes[i]
    state_data = covid[covid["State"] == top_5_states.index[i]].groupby(["Year", "Month"])["COVID-19 Deaths"].sum().reset_index()
    sns.lineplot(data=state_data, x="Month", y="COVID-19 Deaths", hue="Year", marker="o", ax=ax)
    ax.set_title(f"COVID-19 Death Trend in {top_5_states.index[i]}")
    ax.set_xlabel("Month")
    ax.set_ylabel("Total COVID-19 Deaths")
    ax.grid(True)
    ax.legend(title="Year")

# Removing empty subplot if any
fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

Key observations:

There are clear peaks indicating pandemic waves, with different states experiencing them at slightly different times. Some states had multiple surges, showing the impact of different COVID-19 waves. There is a gradual decline in later months, likely due to vaccinations and public health measures.



```python
# Comorbidities Correlation
df_corr = covid[['COVID-19 Deaths', 'Pneumonia Deaths', 'Influenza Deaths', 'Pneumonia, Influenza, or COVID-19 Deaths']]
correlation = df_corr.corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt='.2f')
plt.title("Correlation Between COVID-19, Pneumonia, and Influenza Deaths")
plt.show()
```

Key Insights from Correlation Analysis:
Strong correlation (close to 1) between COVID-19 deaths and total deaths → This confirms that COVID-19 significantly contributed to overall mortality.
High correlation with pneumonia deaths → Many COVID-19 cases involved pneumonia, which aligns with medical findings.
Weaker correlation with influenza deaths → While both are respiratory illnesses, influenza deaths did not always trend with COVID-19 deaths.

```python
# Seasonal trend analysis (COVID-19 deaths by month across all years)
plt.figure(figsize=(12, 6))
sns.lineplot(data=covid.groupby("Month")["COVID-19 Deaths"].sum().reset_index(), x="Month", y="COVID-19 Deaths", marker="o")
plt.title("Seasonal Trends in COVID-19 Deaths (All Years)")
plt.xlabel("Month")
plt.ylabel("Total COVID-19 Deaths")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(True)
plt.show()
```

Seasonal Trend Insights:
COVID-19 deaths peaked during January.
Lower deaths in summer months (March - July), likely due to reduced virus transmission.
These trends align with past observations of respiratory illnesses worsening in colder months.

Conclusion:

Overall Trends
